```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
x=le.fit_transform(np.array(data['protocol_type']))
```

```python
len(x)
print(x)
print(x.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(sparse=False)
x=x.reshape(len(x),1)
x = onehotencoder.fit_transform(x)
print(x.shape)
print(x)
```

```python
li1=[i[0] for i in x]
li2=[i[1] for i in x]
li3=[i[2] for i in x]
```

```python
li1 = pd.Series(li1)
li2 = pd.Series(li2)
li3 = pd.Series(li3)
```

```python
del data['protocol_type']
```

```python
data['protocol_type1']=pd.Series(li1)
data['protocol_type2']=pd.Series(li2)
data['protocol_type3']=pd.Series(li3)
data
```

```python
del data['flag']
```

```python
del data['service']
```

```python
data
```

```python
from sklearn.utils import shuffle
ratio=2/3
train=data[:int(len(data)*ratio)]
test=data[int(len(data)*ratio):]
```

```python
y_train=train['label']
x_train=train
del x_train['label']
y_test=test['label']
x_test=test
del x_test['label']
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1)
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```

```python
clf.score(x_test,y_test)
```
